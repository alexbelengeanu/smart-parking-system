```python
import cv2
from PIL import Image
import numpy as np
```

```python
camera = cv2.VideoCapture(0)
```

```python
result, frame = camera.read()
```

```python
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
frame_rgb = cv2.circle(frame_rgb, (250,100), radius=5, color=(0, 0, 255), thickness=-1)
frame_rgb = frame_rgb[100:420, 0:640]
```

```python
display(Image.fromarray(frame_rgb))
```
