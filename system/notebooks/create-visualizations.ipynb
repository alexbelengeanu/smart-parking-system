```python
import matplotlib.pyplot as plt
import cv2 as cv
```

### Figura 1.3.1

```python
side_view_1 = cv.imread(r'E:\GitHub\smart-parking-system\references\images\side_view_1.jpeg')
side_view_2 = cv.imread(r'E:\GitHub\smart-parking-system\references\images\side_view_2.jpeg')
upper_view = cv.imread(r'E:\GitHub\smart-parking-system\references\images\upper_view.jpeg')
```

```python
side_view_1 = cv.cvtColor(side_view_1, cv.COLOR_BGR2RGB)
side_view_2 = cv.cvtColor(side_view_2, cv.COLOR_BGR2RGB)
upper_view = cv.cvtColor(upper_view, cv.COLOR_BGR2RGB)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 5))

ax[0].axis('off')
ax[0].imshow(side_view_1, aspect='auto')

ax[1].axis('off')
ax[1].imshow(upper_view, aspect='auto')

ax[2].axis('off')
ax[2].imshow(side_view_2, aspect='auto')
plt.savefig(r'E:\GitHub\smart-parking-system\references\images\1-3-1_proiect.jpg', bbox_inches='tight')
```
